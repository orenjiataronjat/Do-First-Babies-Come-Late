```python
from __future__ import print_function, division

import nsfg
```

```python
preg = nsfg.ReadFemPreg()
```

```python
baby = [0,0,0,0,0,0,0,0,0]
weeks = [0,0,0,0,0,0,0,0,0]
preglngth = preg.prglngth[preg.outcome == 1]

for i in range(9):
    baby[i] = preg.prglngth[preg.birthord == i+1]
    weeks[i] = sum(baby[i]) / len(baby[i])
    
    print("Baby: " + 
          str(i) + "\t\t" + 
          "Length of pregnancy in weeks: " +
          str(round(weeks[i], 3)) + 
          "\t\t Number of respondants: " + 
          str(len(baby[i])))
```

```python
import matplotlib.pyplot as plt
```

```python
#plt.scatter(babyNP, weeksNP, color = 'blue')
plt.plot(range(9), weeks)
plt.xlabel('Birth Order')
plt.ylabel('Length of pregnancy')
plt.title('Do first babies come late?')
plt.show()
```
